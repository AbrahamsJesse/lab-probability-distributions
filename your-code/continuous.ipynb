    * It should accept 3 parameters: 
        * `bottom` - the lower boundary of the generated numbers
        * `ceiling` - the upper boundary of the generated numbers
        * `count` - how many numbers to generate
    * It should return an array of uniformly distributed random numbers

1. Call your function with 2 sets of params below:
    * bottom=10, ceiling=15, count=100
    * bottom=10, ceiling=60, count=1,000

1. Plot the uniform distributions generated above using histograms, where x axis is the value and y axis is the count. Let the histogram's number of bins be 10.

Your output should look like below:

![uniform distribution](ud.png)

```python
# your code here
from scipy.stats import uniform
x = uniform.rvs(size=10)
a = 2
b = 3
randoms  = a + (b-a)*x
print(randoms)
```

```python
import matplotlib.pyplot as plt
from scipy.stats import uniform

def generate_uniform_random(bottom, ceiling, count):
    # Generate uniformly distributed random numbers
    x = uniform.rvs(size=count)
    random_numbers = bottom + (ceiling - bottom) * x
    return random_numbers

# Call the function with the provided sets of parameters
random_set1 = generate_uniform_random(bottom=10, ceiling=15, count=100)
random_set2 = generate_uniform_random(bottom=10, ceiling=60, count=1000)

# Plot histograms
plt.figure(figsize=(12, 6))

plt.subplot(1, 2, 1)
plt.hist(random_set1, bins=10, color='blue', alpha=0.7)
plt.title('Uniform Distribution (10, 15), Count=100')
plt.xlabel('Value')
plt.ylabel('Count')

plt.subplot(1, 2, 2)
plt.hist(random_set2, bins=10, color='green', alpha=0.7)
plt.title('Uniform Distribution (10, 60), Count=1000')
plt.xlabel('Value')
plt.ylabel('Count')

plt.show()
```

## Normal Distribution

1. In the same way in the Uniform Distribution challenge, write a function that generates normally distributed random numbers.
1. Generate 1,000 normally distributed numbers with the average of 10 and standard deviation of 1
1. Generate 1,000 normally distributed numbers with the average of 10 and standard deviation of 50
2. Plot the distributions of the data generated.

Expected output:

![normal distribution](nd.png)

```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import norm

def generate_normal_random(mean, std_dev, count):
    # Generate normally distributed random numbers
    normal_numbers = np.random.normal(loc=mean, scale=std_dev, size=count)
    return normal_numbers

# Generate 1,000 normally distributed numbers with mean of 10 and standard deviation of 1
data_set1 = generate_normal_random(mean=10, std_dev=1, count=1000)

# Generate 1,000 normally distributed numbers with mean of 10 and standard deviation of 50
data_set2 = generate_normal_random(mean=10, std_dev=50, count=1000)


############################################################################################
# Plot the distributions
plt.figure(figsize=(12, 6))

plt.subplot(1, 2, 1)
plt.hist(data_set1, bins=30, color='red', alpha=0.7, density=True)
plt.title('Normal Distribution (Mean=10, Std Dev=1)')
plt.xlabel('Value')
plt.ylabel('Probability Density')
plt.grid(True)

plt.subplot(1, 2, 2)
plt.hist(data_set2, bins=30, color='orange', alpha=0.7, density=True)
plt.title('Normal Distribution (Mean=10, Std Dev=50)')
plt.xlabel('Value')
plt.ylabel('Probability Density')
plt.grid(True)

plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

mean1=1
mean2=100
def exponential (mean, size):
    
    return np.random.exponential(scale=mean, size=size)

sequence1 = exponential (mean1, size=1000)
sequence2 = exponential (mean2, size=1000)


```

## Exponential Distribution of Real Data

Suppose that the amount of time one spends in a bank is exponentially distributed with mean as 10 minutes (i.e.  = 1/10). What is the probability that a customer will spend less than fifteen minutes in the bank? 

Write a code in python to solve this problem

```python
# your answer here
# Hint: This is same as saying P(x<15)


lambda_inv = 10/1 
expon_dist = expon (scale=lambda_inv)

print('The probability that a customer will spend LESS than fifteen minutes in the bank is:', expon_dist.cdf(15))
```

What is the probability that the customer will spend more than 15 minutes

```python
# your answer here

#What is the probability that the customer will spend more than 15 minutes
lambda_inv = 10/1 
expon_dist = expon (scale=lambda_inv)

print('The probability that a customer will spend MORE than fifteen minutes in the bank is:', 1- expon_dist.cdf(15))
```
